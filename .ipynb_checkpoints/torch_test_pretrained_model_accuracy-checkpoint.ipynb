```python
class ImageNetDataset(Dataset):
    """Face Landmarks dataset."""

    def __init__(self, csv_file, root_dir, transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.dataframe = pd.read_csv(csv_file,delimiter=" ")
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_name = os.path.join(self.root_dir,self.dataframe.iloc[idx, 0])
        image1=cv2.imread(img_name)
        image = Image.fromarray(image1)
        label = self.dataframe.iloc[idx, 1]
        #image = io.imread(img_name)
        #landmarks = self.df.iloc[idx, 1:]
        #landmarks = np.array([landmarks])
        #landmarks = landmarks.astype('float').reshape(-1, 2)
        #sample = {'image': image , 'ground_truth': g_t}
        
        if self.transform:
            image = self.transform(image)

        return image,label
```

```python
csv_file_1= '/home/charlene/15aug_val_dataset/val.csv'
root_directory = '/home/charlene/15aug_val_dataset'
```

```python
transforms = torchvision.transforms.Compose([
        torchvision.transforms.Resize(size=(224,224)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])
```

```python
imagenet_dataset = ImageNetDataset(csv_file=csv_file_1,root_dir=root_directory,transform=transforms)


```

```python
val_dataloader = DataLoader(imagenet_dataset, batch_size=16,
                        shuffle=False, num_workers=0)
```

```python
#top 5 prediction
from tqdm import tqdm
count=0
total_images=0
for image_batch,label_batch in tqdm(val_dataloader):
    #print(image_batch.shape)
    #total_images+= image_batch.shape[0]
    #print(label_batch)
    image_batch=image_batch.cuda()
    label_batch=label_batch.cuda()
    pred=model(image_batch)
    pred_sort=torch.argsort(pred,dim=1,descending=True)
    #print(pred_sort.shape)
    #print(pred_sort[:,:5])
    top_5_prediction=pred_sort[:,:5]
    testing=torch.nn.functional.one_hot(top_5_prediction, num_classes=1000)
    prediction=torch.sum(testing, dim=1, keepdim=False)
    g_t_one_hot=torch.nn.functional.one_hot(label_batch, num_classes=1000)
    count+=(prediction*g_t_one_hot).sum()
    #print(top_5_prediction)
    #print(torch.argmax(pred,axis=1))
    #count+=(torch.argmax(pred,axis=1)==label_batch).sum()
    
    
```

```python
top_5_accuracy = (count/50000)*100
print(f"Top 5 accuracy for pytorch pretrained model is {top_5_accuracy}%")
```
